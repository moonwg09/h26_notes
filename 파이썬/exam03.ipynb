```python
# 키보드 입력 방법1 
# input() : 키보드 입력받는 함수
num1 = input('첫번째 정수를 입력하세요')
num2 = input('두번째 정수를 입력하세요')

# int(), float(), str() : type 변환함수

num1 = int(num1)  # '34' => 34
num2 = int(num2)

```

```python
# 키보드 입력 방법2
num1 = int(input('첫번째 정수 입력 :'))
num2 = int(input('두번째 정수 입력 : '))
```

```python
# 키보드 입력 방법3
# str = '2026-3-9'
# print(str)
# print(str.split('-'))

str = input('정수 2개를 입력하세요 ex)12 6')        # str = '12 6'
print(str, type(str))
str = str.split(' ')  # ['12', '6']
print(str, type(str)) 
# map(수행할함수명, 집합) : 집합의 모든 데이터에 지정한 함수를 수행하시오
num1, num2 = map(int, str)  # num1, num2 = [12, 6]
```

```python
# str = input('정수 2개를 입력하세요 ex)12 5')
# num1, num2 = map(int, str.split(' '))

num1, num2 = map(int, input('정수 2개를 입력하세요 ex) 12 9').split(' '))
```

```python
# 출력1
print('num1 :' , num1,  'num2 :', num2)
print('두 정수의 합 : ', num1 + num2)
print("두 정수의 차 : ", num1 - num2)
print('두 정수의 곱 : ', num1 * num2)
print('두 정수의 나눗셈 : ', num1 / num2)
print('두 정수의 몫 : ', num1 // num2)
print('두 정수의 나머지 : ', num1 % num2)
```

```python
# 출력2
print('num1 :' , num1,  'num2 :', num2)
print(num1,' + ', num2 , ' = ', (num1 + num2), sep='')
print(num1, '-', num2, '=', num1-num2)
```

```python
# 출력3
# 포맷문자 사용  %d  %f  %s
print('%d + %d = %d' % (num1, num2, num1 + num2))
print('%d %% %d = %d' % (num1, num2, num1 % num2))
```

```python
# 출력4
str = '{0} + {1} = {2}'.format(num1, num2, num1 + num2)
print(str)
print('{} - {} = {}'.format(num1, num2, num1 - num2))
```

```python
# 출력5
print(f'{num1} + {num2} = {num1 + num2}')
```
