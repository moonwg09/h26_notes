open(파일명, 파일모드)           파일 객체 연다  파일모드 'r' 'w'(w모드는 있으면 지우고 다시 만들어줘서 조심해야한다)

read()/write()   read/write

close()          파일 객체 닫는다

```python
file = open('test.txt','w')
file.write("hello")
file.close()
```

```python
file = open("test.txt", 'r') #하드디스크에 없으면 에러가 뜬다
data = file.read()
file.close()
print(data)
```

```python
with open("test.txt", 'r') as file:
    data = file.read()
print(data)
```

```python
# quit가 나올떄까지 문자열을 입력받아 output.txt에 저장한 후,
# output.txt에 저장된 내용을 출력하는 코드를 작성
# output.txt에 저장된 단어의 개수를 출력

with open("output.txt", 'w') as file:
    count = 0
    data = input()
    count +=1
    file.write(data)
    if data == 'quit':
        file.close()

with open("output.txt", 'r') as file:
    data = file.read()
print(data)
```
