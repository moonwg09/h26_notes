```python
with open('test3.txt', 'w') as file:
    print('문자열을 입력하세요. quit 입력시 종료')
    while True:
        msg = input()
        if msg.lower() == "quit":
            break
        file.write(f'{msg}\n')
```

```python
# 비추천 , read로 통채로 들고오면 위험하다(용량이 크면 부하를 가질 수 있다)
with open('test3.txt', 'r') as file:
    data = file.read()
    print(data)
```

```python
# readline() , 한줄씩 가져온다
with open('test3.txt', 'r') as file:
    while True:
        line = file.readline().strip("\n")
        if line == "":                  #eof(end of file) 설정
            break
        #line = line.strip('\n')         #enter까지 문자로 쳐서 수가 늘기 때문에 없애준다
        print(line, len(line))
```

```python
with open('test3.txt', 'r') as file:
    for line in file:
        line = line.strip('\n')
        print(line)
```

```python
with open('test3.txt', 'r') as file:
    lines = file.readlines()
print(lines)

for line in lines:
    print(line.strip('\n'))
```
