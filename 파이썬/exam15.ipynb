```python
# 비밀번호 변경서비스
members = {'aaa':'1111','bbb':'2222','ccc':'3333','ddd':'4444'}

while True:
    print("*** 패스워드 변경 서비스 ***")
    id = input("아이디를 입력하세요 : ")
    if id not in members:
        print(f"아이디[{id}]는 존재하지 않습니다.")
        print("변경서비를 종료합니다")
        break
    else:
        password = input("현재 패스워드를 입력하세요 : ")
        if members.get(id) != password:
            print("패스워드가 올바르지 않습니다.")
            print("변경서비스를 종료합니다.")
            break
        else:
            newpassword = input("변경할 패스워드를 입력하세요 : ")
            members.update([[id,newpassword]]) #members.update({id:newpassword})
            print("패스워드를 변경하였습니다.")
            break


print("<<<회원정보>>>")
for key,value in members.items():
    print(key,value)


```
