```python
# 교집합, 합집합, 차집합, 여집합 연산 수행
a = {1,2,3,4}
b = {3,4,5,6}
c = {5,6,7,8}
print(f'합집합 : {set.union(a,b,c)}')
print(f'합집합 : {a | b}')
print(f'합집합 : {a.union(b)}')
```

```python
a = {1,2,3,4}
b = {3,4,5,6}
c = {5,6,7,8}
print(f'교집합 : {set.intersection(a,b)}')
print(f'교집합 : {a & b}')
print(f'교집합 : {a.intersection(b)}')
```

```python
a = {1,2,3,4}
b = {3,4,5,6}
c = {5,6,7,8}
print(f'차집합 : {set.difference(a,b)}')
print(f'차집합 : {set.difference(b,a)}')
print(f'차집합 : {a - b}')
print(f'차집합 : {b - a}')
print(f'차집합 : {a.difference(b)}')
print(f'차집합 : {b.difference(a)}')
```

```python
a = {1,2,3,4}
b = {3,4,5,6}
c = {5,6,7,8}
print(f'여집합 : {set.symmetric_difference(a,b)}')
print(f'여집합 : {a-b|b-a}')
print(f'여집합 : {a^b}')
print(f'여집합 : {a.symmetric_difference(b)}')
```

```python
a = {1,2,3,4}
b = {3,4,5,6}
# a = set.intersection(a,b)
# a = a.intersection(b)
# a = a & b
# set.intersection_update(a,b)
a &= b
print(a)
print(b)
```

```python
a = {1,2,3,4}
print(a)
# 원소 5 삽입
#a |= {5}
a.update({5})
print(a)
```

```python
# 부분집합, 진부분집합
a = {1,2,3,4}
print(f'부분집합 : {a <= {1,2,3,4}}')
print(f'부분집합 : {a.issubset({1,2,3,4})}')
print(f'진부분집합 : {a < {1,2,3,4}}')
print(f'진부분집합 : {a < {1,2,3,4,5}}')
```

```python
# 집합 사이의 교집합 원소 존재여부 판단(존재X : true, 존재O: false)
a = {1,2,3,4}
print(a.isdisjoint({1,2,5,6}))
print(a.isdisjoint({5,6}))
```

```python
a = {1,2,3,4}
b = a
print(a == {1,2,3,4})
print(a == {4,3,2,1})
print(a == {4,3,2})
print(a != {4,3,2})
print(a is {1,2,3,4})
print(a is a)
print(a is b)
```

#### set 조작함수
- add() : 하나씩만 넣을 수 있다.
- remove() : 원소 삭제, 단 삭제할 원소가 없을 때 예외발생
- discard() : 원소 삭제, 삭제할 원소가 없어도 예외발생X
- pop()
- clear()
- copy() : deep copy

```python
a = {1,2,3,4}
a.add(5)
print(a)
len(a)
```

```python
a.remove(3)
print(a)
```

```python
print(a)
a.discard(2)
print(a)
a.discard(2)
```

```python
a = {1,2,3,4,5}
print(f'pop() : {a.pop()}')
print(a)
print(f'pop() : {a.pop()}')
print(a)
```
