```python
#함수 정의 / 함수 선언
def printStar():
    print('*' * 5)
```

```python
#함수 실행 / 함수 호출
printStar()
print("Hello")
printStar()
print('Hi')
printStar()
```

```python
def printChars(c): #c는 파라미터 혹은 매개변수라고 부른다(argument)
    print(f'{c}' * 5)
```

```python
printChars('*')
print("Hello")
printChars('-')
print('Hi')
printChars('~')
```
