## 선택 정렬(Selection Sort)

```python
#선택 정렬

#리스트는 주소값만 참조하기 때문에 return안해줘도 된다
def selection_sort(A):
    N = len(A)
    for start in range(N-1):
        min_index = select_min(A, start)
        A[start],A[min_index] = A[min_index],A[start]
        print_step(A, start)
    #return A

def select_min(A, start):
    min = A[start]
    min_index = start
    for i in range(start+1, len(A)):
        if min > A[i]: 
            min = A[i]
            min_index = i
    return min_index

def print_step(data, step):
    print(f' step {step} : {data}')
    

data = [5,3,8,4,9,1,6,2,7]
print('Orginal :', data)
selection_sort(data)
print('Sorted  :',data)
    
```

## 삽입 정렬(Insertion Sort)

```python
# 삽입 정렬

def insertion_sort(A):
    N = len(A)
    for cur_index in range(1,N):
        key = A[cur_index]
        i = cur_index - 1            #정렬된 부분의 마지막 데이터부터
        while i >= 0 and A[i] > key: #key보다 큰 데이터인 동안 
            A[i+1] = A[i]            #데이터를 뒤로 한 칸씩 이동
            i -= 1                   #한 칸 앞으로 index 이동
        # 나보다 작은 애를 만난 상태
        A[i+1] = key
        print_step(A,cur_index)
    
def print_step(data, step):
    print(f' step {step} : {data}')
    

data = [5,3,8,4,9,1,6,2,7]
print('Orginal :', data)
insertion_sort(data)
print('Sorted  :',data)
            
```

## 버블 정렬(Bubble Sort)

```python
# 버블 정렬

def bubble_sort(A):
    N = len(A)
    is_changed = False
    for last in range(N-1, 0, -1):
        
        for i in range(last):
            if A[i]>A[i+1]:
                A[i],A[i+1] = A[i+1],A[i]
                is_changed = True
        # swap된 경우가 없으면 = 정렬이 안된 부분이(last 앞부분) 이미 정렬이 되어 있으면 정렬을 멈춤
        if  not is_changed:
            break
        print_step(A, last)
        
def print_step(data, step):
    print(f' step {step} : {data}')
    

data = [5,3,8,4,9,1,6,2,7]
print('Orginal :', data)
bubble_sort(data)
print('Sorted  :',data)
```
