다음의 결과를 보이는 프로그램을 함수로 이용하여 작성하시오
학생 5명의 성적을 입력받아 다음과 같이 출력하는 코드를 작성하시오.

1's 성적 : 83
2's 성적 : 1000
성적은 0~100점사이로만 입력하세요
2's 성적 : 90
3's 성적 : 67
4's 성적 : 99
5's 성적 : 1000
성적은 0~100점사이로만 입력하세요
5's 성적 : 79

-------------------------------------
    번호        성적        학점
-------------------------------------
    1           83          B
    2           90          A
    3           67          D
    4           99          A
    5           79          C
-------------------------------------

총점 : 
평균 : 
최고점 : 
최저점 : 

```python
# 학점 구하는 함수
def getGrade(score):
    if score < 0 or score > 100: 
        return 0
    elif 90 <= score <= 100: grade = 'A'
    elif 80 <= score < 90: grade = 'B'
    elif 70 <= score < 80: grade = 'C'
    elif 60 <= score < 70: grade = 'D'
    else: grade = 'F'
        
    return grade

# 총점 구하는 함수
def getTotal(scorelist):
    total_sum = 0  # 내장 함수 sum과 이름이 겹치지 않게 변경하고, 0으로 초기화
    for num in scorelist:  # 인덱스 대신 리스트의 요소를 직접 순회
        total_sum += num
    return total_sum

# 평균 구하는 함수
def getAverage(total):
    average = total / 5
    return average

# 최고점 구하는 함수
def getMax(scorelist):
    max_val = scorelist[0] # 첫 번째 값으로 초기화
    for num in scorelist:
        if max_val < num:
            max_val = num
    return max_val  # for문이 다 끝난 후 return

# 최저점 구하는 함수
def getMin(scorelist):
    min_val = scorelist[0] # 0이 아닌 첫 번째 값으로 초기화
    for num in scorelist:
        if num < min_val:
            min_val = num
    return min_val  # for문이 다 끝난 후 return


# 메인 실행 부분
scores = [] # 점수를 저장할 빈 리스트 생성

for i in range(1, 6): # 1부터 5까지 반복
    while True: # 올바른 점수를 입력받을 때까지 반복
        num = int(input(f"{i}'s 성적 : "))
        grade = getGrade(num)
        
        if grade == 0:
            print("성적은 0~100점사이로만 입력하세요")
        else:
            scores.append(num) # 정상 점수면 리스트에 추가
            break # while 루프를 빠져나가 다음 학생으로 넘어감

# 전체 출력 테이블 작성
print("\n-------------------------------------")
print("    번호        성적        학점")
print("-------------------------------------")
for i in range(5):
    score = scores[i]
    grade = getGrade(score)
    print(f"    {i+1}          {score}          {grade}")
print("-------------------------------------\n")

# 계산 및 최종 출력
total = getTotal(scores)      # 매개변수로 scores 리스트 전달
average = getAverage(total)   
max_score = getMax(scores)    # 매개변수로 scores 리스트 전달
min_score = getMin(scores)    # 매개변수로 scores 리스트 전달

print(f'총점 : {total}')
print(f'평균 : {average}')
print(f'최고점 : {max_score}')
print(f'최저점 : {min_score}')
```

2. 3월 16일자 과제를 함수로 작성하시오.

```python
import random

# 랜덤 숫자 생성 함수
def getRandomNum():
    return random.randint(1, 100)

randomNum = getRandomNum()
chances = 7
low = 1
high = 100

# 기회가 남아있는 동안 반복
while chances > 0:
    num = int(input(f"{low} - {high}사이의 컴퓨터가 생각하는 숫자를 입력하세요 : "))
    
    # 예외 처리
    if num < low or num > high:
        continue 
        
    chances -= 1 # 기회 차감

    # 비교 로직
    if num == randomNum:
        print("정답입니다.\n")
        break #  
        
    elif num < randomNum:
        print(f"{num}보다 큰수입니다.")
        print(f"기회는 {chances}번 남았습니다.\n")
        low = num
        
    elif num > randomNum:
        print(f"{num}보다 작은수입니다.")
        print(f"기회는 {chances}번 남았습니다.\n")
        high = num 

#  chances가 0이 된 경우
if chances == 0 and num != randomNum:
    print("기회를 다 소진하셨습니다.")
    print(f"정답은 [{randomNum}]입니다.")
```

```python
def computerGame():
    low = 1
    high = 100
    chances = 6

    print("1부터 100 사이의 숫자를 마음속으로 하나 생각해주세요!\n")

    while chances > 0:
       
        guess = (low + high) // 2
        
        print(f"생각하는 숫자가 [{guess}]입니까?")
        answer = int(input("정답은 0  작은수면 1  큰수면 2를 입력하세요 :  "))
        
        if answer == 0:
            print("컴퓨터가 정답을 맞췄습니다!")
            return  # 정답을 맞추면 즉시 함수를 종료합니다.
            
        elif answer == 1:
            # 사용자가 생각한 숫자가 컴퓨터의 추측값(guess)보다 '작은 수'인 경우
            high = guess - 1
            
        elif answer == 2:
            # 사용자가 생각한 숫자가 컴퓨터의 추측값(guess)보다 '큰 수'인 경우
            low = guess + 1
            
        else:
            print("잘못 입력하셨습니다. 0, 1, 2 중 하나를 입력해주세요.")
            continue # 잘못된 입력은 기회를 차감하지 않고 다시 묻습니다.
            
        chances -= 1
        print() # 보기 편하게 한 줄 띄우기

    # while문이 끝날 때까지 return을 만나지 못했다면 기회를 모두 소진한 것입니다.
    print("6번 안에 맞추지 못했습니다.")
    real_answer = input("생각하셨던 정답을 알려주세요 : ")
    print(f"아하, 정답은 [{real_answer}]였군요!")

# 프로그램 실행
computerGame()
```
