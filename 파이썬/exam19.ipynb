```python
def print_book(title, writer, price):
    print("제목 :", title)
    print("지은이 :", writer)
    print("가격 :", price)
```

```python
print_book('홍길동전', '허균', 19000)
print_book(title='홍길동전', writer='허균', price=19000)
```

```python
book = {'title' : '홍길동전', 'writer' : '허균', 'price' : 19000}
print(book)
print_book(**book)  #딕셔너리에서 *한개면 key *두개면 value값을 출력한다
print_book(*book)
```

```python
def prn_book(**args):
    #print(args)
    for key, value in args.items():
        print(f'{key} : {value}')
```

```python
prn_book(title='홍길동전', writer='허균', price=19000)
prn_book(title='신데렐라', price=14000)
prn_book(**book)
```

문자열 함수
- replace(x,y) : x에 해당하는 문자열을 y로 대체
- split() 
- join()
- upper() : 대문자로 바꾸기
- lower() : 소문자로 바꾸기
- lstrip()
- rstrip()
- strip() : 좌우 공백을 제거
- center() : 가운데 정렬
- zfill() : 왼쪽 공백을 0으로 채우기
- find(x) : x 문자열 위치 검색
- rfind() : 오른쪽부터 검색
- index() : 없으면 error가 뜬다 (find는 없어도 error가 안뜸)
- count() 

```python
print(str)
str01 = str.replace('l', 'rr')  # 문자열은 상수이기 때문에 바뀌지 않아서 return값으로 받아줘서 바꿔줘야한다
print(str01)
```

```python
str02 = ':'.join('홍길동전 허균')
print(str02)
```

```python
print('Hello world!!!'.upper())
print('Hello world!!!'.lower())
```

```python
print("[%s]" % str.center(20))
print("[%s]" % str.ljust(20))
print("[%s]" % str.zfill(20))
```

```python
print(str.find('l'))
print(str.index('l'))
print(str.find('lo'))
print(str.index('lo'))
print(str.find('p'))
print(str.count('l'))
```
