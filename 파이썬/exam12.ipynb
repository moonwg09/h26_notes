```python
# 호출자가 2개의 값을 전달하면 두수의 합을 구해서 호출자에게 전달하는 getSum 함수를 정의
def getSum(a,b):
    s = a+b
    return s
```

```python
def getTotal(a=0,b=0,c=0):
    sum = a + b + c
    return sum
```
