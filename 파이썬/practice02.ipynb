1번 문제

```python
a, b, c, d = map(int, input("숫자 4개 입력: ").split(" ")) # 4개의 숫자를 입력받아서 각각 int형으로 자료형을 바꿔준다.

max = 0 # 가장 큰 수를 max로 지정해서 옮겨주는 역할을 한다.

if a>b :        #swap
    max = a
else:
    max = b

if max<c:
    max = c

if max<d:
    max = d

print("12, 6, 34, 9 중 가장 큰수 : %d" % max)  # 출력
```

3번 문제

```python
score = int(input("성적 입력 : "))  #성적을 입력받는 정수를 만든다.

if 90<=score and score<100:     # and로 조건을 둘다 만족하면 실행한다.
    grade = 'A'
elif 80<=score and score<90:
    grade = 'B'
elif 70<=score and score<80:
    grade = 'C'
elif 60<=score and score<70:
    grade = 'D'
else:
    grade = 'F'

print("성적 입력 : %d" % score)
print("%d점의 학점은 %s" % (score, grade))
```

5번 문제

```python
price = int(input("물건값을 입력하시오 : "))
money = int(input('지불한 돈의 액수를 입력하시오 : '))

change = money - price      #거스름돈을 구하는 방법

if change >= 1000:                      # 거스름돈의 몫을 a, 나머지를 다시 change변수에 넣어서 다음 조건식을 이행한다.
    a = change // 1000  
    change = (change - (1000 * a))
else:
    a = 0                               # 조건식에 맞지 않으면 몫이 없기때문에 0값을 넣어준다.
if change>=500:
    b = change // 500
    change = change - (500 * b)
else:
    b = 0
if change>=100:
    c = change // 100
    change = change - (100 * c)
else:
    c = 0
if change>=50:
    d = change // 50
    change = change - (50 * d)
else:
    d = 0
if change>=10:
    k = change // 10
else:
    k = 0

print("거스름돈 : %d원" % (money-price))        #출력
print("1000원 : %d개" % a)
print("500원 : %d개" % b)
print("100원 : %d개" % c)
print("50원 : %d개" % d)
print("10원 : %d개" % k)
```
