### 01. 약수, 공약수, 최대공약수 구하기

```python
# 약수

def get_divisors(N):
    divisor_list = []
    for num in range(1,N+1):
        if N % num == 0:
            divisor_list.append(num)
    return divisor_list

number = int(input("숫자 입력: "))
print("약수 : ", get_divisors(number))
```

```python
number = int(input("숫자 입력: "))
divisor = get_divisors(number)
print("약수 : ", " ".join(map(str,divisor)))
```

```python
# 공약수

def get_common_divisors(M,N):
    if M > N : M, N = N, M
    cdivisor_list = []
    for num in range(1, M+1):
        if M % num == 0 and N % num == 0:
            cdivisor_list.append(num)
    return cdivisor_list

cdivisors = get_common_divisors(12,18)
print("공약수: ", cdivisors)
```

```python
num1, num2 = map(int, input().split())
divisors = get_common_divisors(num1,num2)
print("공약수 : ", " ".join(map(str,divisors)))
```

```python
# 최대 공약수 구하기

def get_gcd(M, N):
    if M > N : M, N = N, M

    for num in range(M,0,-1):
        if M % num == 0 and N % num == 0:
            return num

gcd = get_gcd(12,18)
print("최대공약수 : ", gcd)
```

```python
num1, num2 = map(int, input().split())
gcd = get_gcd(num1,num2)
print("최대공약수 : ",gcd)
```

### 02. 배수, 공배수, 최대공배수 구하기

```python
# 최소공배수

def get_lcm(M, N):
    if M > N : M,N = N,M

    i = 1
    while True:
        multiple = N * i
        if multiple % M == 0:
            return multiple
        i+=1

lcm = get_lcm(9,4)
print("최소공배수 : ", lcm)
```

## 03. 소수(Prime Number) 구하기

```python
# 소수

def is_prime_number(N):
    if N == 2: return True
    for i in range(2,N):
        if N % i == 0:
            return False
    return True

print(is_prime_number(2))
```

```python
# 소수 알고리즘 3가지 풀이

# 알고리즘 개선하기(1)
def is_prime_number(N):
    while True:
        if N % 2 == 0:
            return False

        else:
            for i in range(3,N,2):
                if N % i == 0:
                    return False
            return True
print(is_prime_number(5))
```

```python
# 알고리즘 개선하기(2)
#2부터 N/2까지 나누어떨어지는 수가 없다면, 그보다 큰 수들 중에서도 약수는 절대 존재할 수 없습니다.
def is_prime_number(N):
    for i in range(2, N//2):
        if N % i == 0:
            return False
    return True

print(is_prime_number(13))
```

```python
# 알고리즘 개선하기(3)

def is_prime_number(N):
    primenumber = 0
    for i in range(2,N):
        for j in range(1,i):
            if i % j == 0:
                primenumber = i
                break
        if N % primenumber == 0:
            return False
    return True
    
print(is_prime_number(20))
```

## 04. 진법 변환 (Carninality)

```python
def convert_base(N, base):
    # 데이터 정의
    digits = '0123456789ABCDEFGHIJKLMNOPQRSTUVWXYZ'

    # 알고리즘
    result = ""

    while N > 0:
        #할 일 먼저 하고
        remainder = N % base
        result = digits[remainder] + result
        # 조건에 해당하는 값 수정하기
        N //= base
        
    return result
N, base = map(int, input("숫자와 base 입력:").split())
print(f'{N}의 {base}진수 : {convert_base(N, base)}')
```

## 04. Two Pointer

```python
#i번째 수부터 j번째 수까지의 합 A[i] + A[i+1] + … + A[j-1] + A[j]가 M이 되는 경우의 수를 구하는 프로그램


#M = 5
#N = 10
#A = [1, 2, 3, 4, 2, 5, 3, 1, 1, 2]

N, M = map(int,input().split())
A = list(map(int,input().split()))

# 정답 데이터 초기화
count = 0

# 알고리즘의 시작값 초기화
start = end = 0
sum = A[0]

while end < N:
    # 할 일 하기
    if sum == M: count += 1
    
    # 다음을 위한 조건 데이터 수정하기
    if sum > M:
        sum -= A[start]
        start += 1
    else:
        end += 1
        if end < N: sum += A[end]

print(count)
```
