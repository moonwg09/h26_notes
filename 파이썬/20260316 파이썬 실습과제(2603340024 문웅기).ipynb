1. 컴퓨터가 1 ~ 100 사이의 숫자를 알고 있다. 사용자가 7번의 기회안에 그 숫자를 맞추는 프로그램을 작성하시고.
   random 모듈을 공부하여 작성

  컴퓨터가 생각하는 숫자가 28이라면,

  1 - 100사이의 컴퓨터가 생각하는 숫자를 입력하세요 : 89
  89보다 작은수입니다.
  기회는 6번 남았습니다
 
  1 - 88사이의 컴퓨터가 생각하는 숫자를 입력하세요 : 25
  25보다 큰수입니다
  기회는 5번 남았습니다

  26 - 88사이의 컴퓨터가 생각하는 숫자를 입력하세요 : 90
  26 - 88사이의 컴퓨터가 생각하는 숫자를 입력하세요 : 11
  26 - 88사이의 컴퓨터가 생각하는 숫자를 입력하세요 : 50
  50보다 작은수입니다
  기회는 4번 남았습니다

  26 - 49사이의 컴퓨터가 생각하는 숫자를 입력하세요 : 30
  30보다 작은 수입니다.
  기회는 3번 남았습니다

  26 - 29사이의 컴퓨터가 생각하는 숫자를 입력하세요 : 28
  정답입니다.

  또는

   26 - 29사이의 컴퓨터가 생각하는 숫자를 입력하세요 : 26
  26보다 큰수입니다. 
  기회는 0번 남았습니다
  기회를 다 소진하셨습니다.
  정답은 [28]입니다.


```python
import random
count = 7  # 가능 횟수를 생성
comnum = random.randint(1,100)  #랜덤 숫자 생성
low = 1
high = 100
num = 0

for i in range(7):
    num = int(input(f"{low}~{high}사이의 컴퓨터가 생각하는 숫자를 입력하세요 : "))
    # 랜덤숫자와 사용자가 선택한 숫자를 비교해서 출력
    if num < comnum:
        count -= 1
        low = num
        print(f'{num}보다 큰수입니다')
        if count == 0:
            print(f'기회를 다 소진하셨습니다.')
            print(f'정답은 [{comnum}]입니다.')
        else:
            print(f'기회는 {count}번 남았습니다')
    elif num > comnum:
        count -= 1
        high = num
        print(f"{num}보다 작은수입니다")
        if count == 0:
            print(f"기회를 다 소진하셨습니다.")
            print(f"정답은 [{comnum}]입니다.")
        else:
            print(f'기회는 {count}번 남았습니다')
    else:  # 사용자가 고른 숫자와 컴퓨터가 랜덤으로 낸 숫자가 같을 때
        print(f"정답입니다!")

```
