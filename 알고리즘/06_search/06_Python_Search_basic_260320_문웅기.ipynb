## 이진 탐색(Binary Search)

```python
# 이진 탐색
def binarty_search(A,key):
    low = 0
    high = len(A)-1

    while low<=high:
        middle = (low+high)//2
        print(f'{low:2}, {middle:2}, {high:2}')
        if key == A[middle]:
            return middle
        if key < A[middle]:
            high = middle -1
        else:
            low = middle + 1
    return -1

data = [1,2,3,4,5,6,7,8,9]
index = binarty_search(data, 8)
print(index, data[index])
```
