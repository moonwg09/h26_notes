# list 함수
- append : 원소 추가
- insert : 특정위치에 원소 추가
- pop    : 마지막 원소 또는 특정 위치의 원소 삭제
- remove : 특정값을 가진 원소 삭제
- extend : 두개의 리스트를 한개의 리스트로 확장
- index  : 특정값의 원소 위치
- count  : 리스트 내 특정값의 개수
- clear  : 전체 데이터 삭제

```python
a = [10,20,30]
print(a, id(a))
```

```python
a.append(40)
print(a, id(a))
```

```python
b = [100,200]
# a.append(b)
a = a + b  # 주소값이 바뀐다(새로운 배열을 만들기 때문에)
print(a, id(a))
```

```python
a.extend(b) # 주소값이 바뀐다(새로운 배열을 만들기 때문에)
print(a, id(a))
```

```python
a.insert(0, 55)
a
```

```python
a.insert(4, 'A')
a
```

```python
a = [10,20,30,40,50]
# a[2] = 1000
#a[2:2] = [1000, 2000]
a[2:3] = [1000] #[2,3]인 부분을 슬라이스해서 빼서 오른쪽에 있는 배열을 집어넣는 방법(여러 개의 값이 있어도 그래도 다 들어간다)
a
```

```python
# delNum = a.pop()
delNum = a.pop(3)
print(delNum)
print(a)
```

```python
a = [10,20,30,40,20,40,60]
a.remove(20)
a
```

```python
delNum = 1000
if a.count(delNum):
    a.remove(delNum)
print(a)
```

```python
a = [10,20,30,40,50]
delIndex = 5
if delIndex >=0 and delIndex < len(a):
    a.pop(delIndex)
print(a)
```

```python
a = [10,20,30,40,50]
print(a)
a.reverse()
print(a)
```

```python
a = [10,20,40,50,15,66]
print(a)
a.sort() # b = sorted(a) , sorted(a, reverse=True) => 내림차순 정령
print(a)
```

```python
a = ['boat', 'banana', 'zoo', 'apple']
print(a)
print(sorted(a))
```

```python
a = [10,20,40,50,15,66]  #문자 sorted는 값이 달라질수 있다
a = list(map(str, a))
print(a)
print(sorted(a))
```

```python
a = [10,20,30,40,50]
#전체 데이터 삭제
#while len(a):
#    a.pop()
#a.clear()
del a[:]
print(a)
```

```python
a = [10, 20, 30, 40, 50]
# b = a  # b는 a의 주소만 복사 shallow copy
b = a[:] # deep copy

##  ==  is
print(f'a == b : {a == b}') #값만 같은지 물어보는 방법
print(f'a is b : {a is b}') #주소값까지 같은지 물어보는 방법

a[2] = 1000

print(a, id(a))
print(b, id(b))
```
