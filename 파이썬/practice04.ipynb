1. 2 ~ 100사이의 소수를 출력하는 코드를 작성 (단, 한라인에 5개씩 출력)

2 3 5 7 11
13 17 19 23
...
97


```python
count = 0  # 소수가 몇 개 출력되었는지 세는 변수

for num in range(2, 101):
    is_prime = True
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            is_prime = False
            break
    
    if is_prime:
        # 소수를 출력 (end=" "를 쓰면 줄바꿈 대신 띄어쓰기를 합니다)
        print(num, end="\t") 
        count += 1  # 소수를 찾았으니 개수를 1 더합니다.
        
        # 만약 4개를 다 채웠다면?
        if count % 5 == 0:
            print()  # 줄바꿈을 실행합니다.
```

2. 시작단을 입력하고, 종료단을 입력하면 그사이의 구구단을 출력하는 코드 작성

    시작단 : 2  9
    종료단 : 9  2

*** 2단 ***
2 * 1 = 2
...
2 * 9 = 18

*** 3단 ***


```python
sdan = int(input("시작단을 입력하세요 : "))
fdan = int(input("종료단을 입력하세요 : "))

if sdan <= fdan:     # 시작단이 종료단보다 작을 경우 ( 시작단 : 2, 종료단 : 9)
    for i in range(sdan, fdan+1):
        print(f"*** {i}단 ***")
        for j in range(1,10):
            print(f"{i} * {j} = {i*j}")
else:
    for i in range(sdan, fdan-1, -1):
        print(f"*** {i}단 ***")
        for j in range(1,10):
            print(f"{i} * {j} = {i*j}")
```

3. 구구단을 출력하는데 한라인에 3단씩 출력

*** 2단 ***     *** 3단 ***    *** 4단 ***

*** 5단 ***

for dan in range(2, 10): 로 시작해야한다.(반복문을 최대한 줄여서 해보자 2,3개로)

```python
# [2, 3, 4단 세트]
for dan in range(2, 5):
    print(f"*** {dan}단 ***", end="\t\t")
print()

for i in range(1, 10):
    for dan in range(2, 5):
        print(f"{dan} * {i} = {dan*i}", end="\t\t")
    print()

print("\n") # 세트 사이 간격을 위해 빈 줄 추가

# [5, 6, 7단 세트] - 이런 식으로 반복하면 됩니다.
for dan in range(5, 8):
    print(f"*** {dan}단 ***", end="\t\t")
print()

for i in range(1, 10):
    for dan in range(5, 8):
        print(f"{dan} * {i} = {dan*i}", end="\t\t")
    print()

print("\n")

# [8, 9단 세트] - 이런 식으로 반복하면 됩니다.
for dan in range(8, 10):
    print(f"*** {dan}단 ***", end="\t\t")
print()

for i in range(1, 10):
    for dan in range(8, 10):
        print(f"{dan} * {i} = {dan*i}", end="\t\t")
    print()
    
```

33이면 짝짝



```python
num = int(input("10이상 100미만의 숫자를 입력하세요 : "))

for i in range(1,num):
    str_i = str(i)   #문자열로 저장한다('13')

    if ('3' in str_i) or ('6' in str_i) or ('9' in str_i):   #문자에 3,6,9가 들어있으면 출력
        print("짝!")
        if (('3' in str_i) or ('6' in str_i) or ('9' in str_i)) and i % 10 == 0: #문자에 3,6,9가 들어있으면서 10의 배수인경우 출력
            print("짝!" + "뽀"*(i//10) + "숑")
    elif i % 10 == 0 :   # 10의 배수일 때만 몫만큼 "뽀"를 붙여 출력                           
        print("뽀" * (i // 10) + "숑")
        if i % 30 == 0:
            print("짝!" + "뽀" * (i // 10) + "숑")
    else:
      print(i)

        
```
