### 딕셔너리 내장 함수
- setdefault() : key 존재하면 아무것도안되지만 key 존재하지않으면 데이터 삽입
- update() : key 존재하면 value 수정, key 존재하지 않으면 새로운 데이터 삽입
- pop()
- popitem()
- clear()
- get(key) : key에 대한 value 값
- items()
- keys()
- values()
- dict.fromkeys()

```python
x = {'a':10,'b':20,'c':30,'d':40}
x
```

```python
x.setdefault('e',50)
print(x)
x.setdefault('f')
print(x)
```

```python
# x['b']= 30
x.update(b=300)  #key가 문자열인 경우에만 update를 사용할 수 있다, 정수형은 update사용 X
x.update({1:'one'})
x.update([[2,"two"]])
print(x)
```

```python
if 'f' in x:
    x.pop('f')
else: print("존재하지 않음")
print(x)
```

```python
print(x)
data = x.popitem()
print(x)
print(data)
```

```python
print(x.get('d'))

key = 'd'
print(x.get(key), x[key])
```

```python
print(x.items())
print(x.keys())
print(x.values())
```

```python
a = ['a','b','c','d']
b = [None for i in a]
print(a)
print(b)
x = dict(zip(a,b))
x
```

```python
a = ['a','b','c','d']
x = {}
for key in a :
    x.setdefault(key)
print(x)
```

```python
y = dict.fromkeys(a)
y = dict.fromkeys(a,0) # dict(zip(['a','b','c','d'], [0,0,0,0]))
y
```

```python
x = {key:None for key in a}
x
```

```python
data = 100
x = {key : data for key,value in dict.fromkeys(a).items()}
x
```

```python
x = {'a': 100, 'b': 200, 'c': 300, 'd':400}
x.keys()
x.values()
y = dict(zip(x.values(), x.keys()))
print(y)
y = {value:key for key,value in x.items()}
print(y)
```

```python
# 딕셔너리 x에서 300값을 가지는 key,value를 삭제
for key in x.keys():
    if x[key] == 300:
        break
print(key)
data = x.pop(key)
print(key, data)
print(x)
```

```python
x = {'a': 100, 'b': 200, 'c': 300, 'd':400,'e':300}

x = {key:value for key, value in x.items() if value != 300}
x
```
