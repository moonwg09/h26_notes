1번 문제 : 학생 5명의 성적을 입력받아 다음과 같이 출력하는 코드를 작성

    1's 학생의 성적 입력 : 80
    80점의 학점은 B다.
    
    2's 학생의 성적 입력 : 1000
    error
    
    3's 학생의 성적 입력 : 91
    91점의 학점은 A다
    
    4's 학생의 성적 입력 : 1000
    error
    
    5's 학생의 성적 입력 : 69
    69점의 학점은 D다.

    5명의 학생 입력 중 2번 에러발생
    총점 : 80 + 91 _ 69
    평균 : .XX점(5명으로 나눠야한다.)

```python
count = 0 # 에러 발생 카운트
total_score = 0  # 총합을 초기화 해놓는다.

for i in range(1,6):  # 1부터 6까지 반복한다.
    
    point = int(input(f"{i}'s 학생의 성적을 입력하세요 : "))

    if point>100:
        print(f"{i}'s 학생의 성적 입력 : {point}")
        print("error", end="\n")
        count += 1
    else :   # 입력받은 성적이 100이하인 경우

        total_score += point

        print(f"{i}'s 학생의 성적 입력 : {point}") # 성적 처리 if문
        if point>=90:
            grade = 'A'
        elif point>=80:
            grade = 'B'
        elif point>=70:
            grade = 'C'
        elif point>=60:
            grade = 'D'
        else:
            grade = 'F'
        print(f"{point}점의 학점은 {grade}이다",end="\n")

# 출력
        
print("\n")
print(f"5명의 학생 입력 중 {count}번 에러발생")
print(f"총점 : {total_score}점")
print("평균 : %.2f점" % (total_score/5))



```

2. 하나의 정수를 입력받아 정수의 약수를 출력하는 코드를 작성
    2~100 사이의 정수 입력 : 45
    정수 45의 약수 출력
    1 3 5 9 15 45

    2 ~ 100 사이의 정수 입력 : 1000
    2 ~ 100 사이의 정수 입력 : 7
    정수 7의 약수 출력
    1 7


```python
num = int(input("2~100 사이의 정수 입력 : ")) #정수값을 입력받는다

print("정수 %d의 약수 출력" % num) 
for i in range(1,num+1):   # 약수 출력하는 for문을 작성
    if num % i == 0:
        print(i, end=" ")
```

3. 3개의 정수를 입력받아 그 정수가 소수인지 판단하는 코드를 작성
   첫번째 정수 : 9
   9는 소수가 아닙니다.
   
   두번째 정수 : 29
   29는 소수입니다.
   
   세번째 정수 : 51
   51는 소수가 아닙니다.


```python
# 3개의 정수를 입력받기 위해 순서 리스트를 만듭니다.
order = ["첫번째", "두번째", "세번째"]

for i in range(3):
    # 1. 정수 입력 받기
    num = int(input(f"{order[i]} 정수 : "))
    
    # 2. 소수 판별 로직
    is_prime = True # 초기값
    
    if num < 2:
        is_prime = False # 1은 제외
    else:
        # 2부터 해당 숫자 직전까지 나누어 봅니다.
        for j in range(2, num):
            if num % j == 0:
                is_prime = False # 하나라도 나누어지면 소수가 아님
                break # 빠져나가기
    
    # 3. 결과 출력
    if is_prime:
        print(f"{num}는 소수입니다.\n")
    else:
        print(f"{num}는 소수가 아닙니다.\n")
```
