1~100사이의 3,5의 공배수 출력

```python
a = set()
b = set()

for i in range(1,101):
    if i % 3 == 0:
        a.add(i)
    if i % 5 == 0:
        b.add(i)

print(a&b)
```

```python
nums3 = {i for i in range(3,101,3)}
nums5 = {i for i in range(5,101,5)}
multiple = set(nums3 & nums5)
print(multiple)
print(min(multiple))
```
