```python
# binary tree traversal(boj 1991)

num_node = int(input())

#트리 생성하기 using dictionary
tree = {} #dict()
for _ in range(num_node):
    data, left, right = input().strip().split()
    tree[data] = [left,right]

def preorder(data):
    if data == '.' : return
    print(data, end = '') # 자기 할 일
    preorder(tree[data][0]) # left subtree
    preorder(tree[data][1]) # right subtree

def inorder(data):+
    if data == '.' : return
    inorder(tree[data][0]) # left subtree
    print(data, end = '') # 자기 할 일
    inorder(tree[data][1]) # right subtree

def postorder(data):
    if data == '.' : return
    postorder(tree[data][0]) # left subtree
    postorder(tree[data][1]) # right subtree
    print(data, end = '') # 자기 할 일
    
preorder('A')
print()
inorder('A')
print()
postorder('A')
```

```python
# min heap
import heapq

heap = []
heapq.heappush(heap, 4)
heapq.heappush(heap, 1)
heapq.heappush(heap, 7)
heapq.heappush(heap, 3)

print(heap)

smallest = heapq.heappop(heap)
print(smallest)
print(heap)
```

```python
# max heap
max_heap = []
heapq.heappush(max_heap, -4)
heapq.heappush(max_heap, -2)
heapq.heappush(max_heap, -7)
heapq.heappush(max_heap, -1)

largest = -heapq.heappop(max_heap)
print(largest)
print(max_heap)
```

```python
# min heap(boj 1927)
import heapq
heap = []

n = int(input())
for _ in range(n):
    num = int(input())
    if num == 0:
        if len(heap) == 0:
            print(0)
        else:
            print(heapq.heappop(heap))
    else:
        heapq.heappush(heap,num)
```

```python
# max heap(boj 11279)
import heapq
heap = []

n = int(input())
for _ in range(n):
    num = int(input())
    if num == 0:
        if len(heap) == 0:
            print(0)
        else:
            print(-heapq.heappop(heap))
    else:
        heapq.heappush(heap,-num)
```
