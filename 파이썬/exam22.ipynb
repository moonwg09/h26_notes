```python
import pickle
```

```python
a = [1,2,3,4,5]

# 원래는 이렇게 하지만
# with open("a.txt", "w") as file:
#     for data in a:
#         file.write(data)

with open("test01.pickle", "wb") as file:  #wb : byte형태로 write
    # pickling
    pickle.dump(a, file)
print('파일저장 완료')
```

```python
with open('test01.pickle', 'rb') as file:
    b = pickle.load(file)
print('로드 완료')
print(b,type(b))
```

```python
def getTotal(*args):
    s = 0
    for data in args:
        s += data
    return s
f = getTotal
print(f(1,2,3,4))
```

```python
def prnTotal(*args):
    s = getTotal(*args)
    print(f'{args} 총합 : {s}')
```

```python
with open('test02.pickle', 'wb') as file:
    pickle.dump(getTotal, file) 
    pickle.dump(prnTotal, file)
print('파일저장 완료...')
```

```python
with open('test02.pickle', 'rb') as file:
    func01 = pickle.load(file)
    func02 = pickle.load(file)
print('파일로드 완료...')
print(type(func01), type(func02), func01(1,2,3))
```

```python
print(func01(*[10,20,30,40,50]))
func02(*[10,20,30,40,50])
```
